# Meroq API Service Foundation

This notebook documents basic local API usage for Release 1.4.0. Start the API with `python scripts/run_api.py --reload` before running request examples.


```python
import requests
BASE_URL = 'http://127.0.0.1:8000'
requests.get(f'{BASE_URL}/health').json()
```

```python
payload = {
    'ticker': 'AAPL',
    'period': '5y',
    'interval': '1d',
    'model_name': 'xgboost',
    'include_news': False,
    'include_risk': False,
}
requests.post(f'{BASE_URL}/analysis/ticker', json=payload).json()['summary']
```
