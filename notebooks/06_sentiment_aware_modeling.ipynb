# Sentiment-Aware Modeling

This notebook audits whether lagged daily sentiment features are ready to be used as model inputs. It is intentionally conservative because recent headlines alone are not enough for a reliable historical sentiment model.

```python
from src.data_loader import fetch_price_data
from src.features import build_model_frame
from src.news_sentiment import fetch_news_for_ticker, analyze_news_sentiment
from src.sentiment_features import aggregate_daily_sentiment
from src.sentiment_modeling import analyze_sentiment_modeling_readiness, compare_base_vs_sentiment_simple_split
```

```python
ticker = 'AAPL'
prices = fetch_price_data(ticker, period='5y', interval='1d')
model_frame = build_model_frame(prices)
news_df, meta = fetch_news_for_ticker(ticker, source='all_configured', max_items=30, use_cache=True)
sentiment_df = analyze_news_sentiment(news_df, engine='lightweight')
daily_sentiment = aggregate_daily_sentiment(sentiment_df, ticker=ticker)
analyze_sentiment_modeling_readiness(model_frame, daily_sentiment).as_dict()
```

```python
result = compare_base_vs_sentiment_simple_split(model_frame, daily_sentiment, model_name='xgboost')
result['comparison']
```
