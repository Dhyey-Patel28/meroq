# Service Layer Research

This notebook shows how to call Meroq's single-ticker analysis service without Streamlit. It is useful for testing, API migration, and batch research.


```python
from src.services import SingleTickerAnalysisRequest, run_single_ticker_analysis, single_ticker_summary_frame
```

```python
request = SingleTickerAnalysisRequest(
    ticker='AAPL',
    period='5y',
    interval='1d',
    model_name='xgboost',
    include_risk=True,
    include_news=True,
    news_source='all_configured',
    sentiment_engine='lightweight',
)
result = run_single_ticker_analysis(request)
single_ticker_summary_frame(result)
```
